```python
# Dependencies
import requests
import pandas as pd
from datetime import datetime
import time
```

```python
# Performing a GET request
url = f"https://data.cityofchicago.org/resource/6zsd-86xi.json?$where=date between '2018-01-01T12:00:00' and '2020-12-31T23:59:59'&$limit=999999"
response = requests.get(url, headers={'X-App-Token': 'TOKEN'}).json()
```

```python
# Storing the retrieved data in a DataFrame
crime_data = pd.DataFrame(response)
```

```python
crime_data.info()
```

```python
crime_data.head()
```

```python
crime_data.columns.to_list()
```

```python
crime_data2 = crime_data[["date", "primary_type"]]
```

```python
crime_data2
```

```python
crime_data2["month"] = pd.to_datetime(crime_data2.date).dt.strftime("%b")
```

```python
crime_data2
```

```python
index_types = crime_data2[(crime_data2['primary_type'] == "RITUALISM") | (crime_data2['primary_type'] == "NON-CRIMINAL (SUBJECT SPECIFIED)") | (crime_data2['primary_type'] == "NON-CRIMINAL")].index
```

```python
crime_data2.drop(index_types, inplace =True)
```

```python
crime_data2['primary_type'].replace("CRIM SEXUAL ASSAULT", "CRIMINAL SEXUAL ASSAULT", inplace=True)
```

```python
crime_data2['primary_type'].value_counts()
```

```python
crime_data_grouped = crime_data2.groupby(['month', 'primary_type']).size().reset_index(name='counts').sort_values(by=['month'])
```

```python
crime_data_grouped['month_num'] = pd.DatetimeIndex(pd.to_datetime(crime_data_grouped['month'], format='%b')).month
```

```python
crime_data_grouped.sort_values(by=['month_num', 'counts'], inplace=True)
```

```python
crime_data_grouped['month'] = crime_data_grouped['month'].str.upper()
```

```python
# crime_data_grouped['normalized'] = (crime_data_grouped['counts']-crime_data_grouped['counts'].min())/(crime_data_grouped['counts'].max()-crime_data_grouped['counts'].min())
```

```python
crime_data_grouped.drop(columns=['month_num'], inplace=True)
```

```python
crime_data_grouped.reset_index(drop=True, inplace=True)
```
